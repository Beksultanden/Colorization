```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def create_professional_vectorscope(image_bgr, title='Vectorscope', size=800):
    """
          
    """
    #  BGR  HSV
    hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
    
    #   H: [0,179]  [0,360]
    h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
    s = hsv[:, :, 1].flatten() / 255.0
    v = hsv[:, :, 2].flatten() / 255.0
    
    #       
    valid_mask = (s > 0.05) & (v > 0.1)
    h_filtered = h[valid_mask]
    s_filtered = s[valid_mask]
    v_filtered = v[valid_mask]
    
    #  
    theta = np.deg2rad(h_filtered)
    x = s_filtered * np.cos(theta)
    y = s_filtered * np.sin(theta)
    
    #     
    N = size
    center = N // 2
    radius = center - 20  #   
    
    #   
    y_coords, x_coords = np.meshgrid(np.arange(N), np.arange(N), indexing='ij')
    dx = x_coords - center
    dy = y_coords - center
    distances = np.sqrt(dx**2 + dy**2)
    
    #   
    circle_mask = distances <= radius
    
    #   ( ,   )
    angles_rad = np.arctan2(dy, dx)
    angles_deg = np.degrees(angles_rad)
    angles_deg = (angles_deg + 360) % 360  #   [0, 360)
    
    #   ()
    normalized_radii = np.minimum(distances / radius, 1.0)
    
    #  HSV 
    hsv_wheel = np.zeros((N, N, 3), dtype=np.float32)
    
    #    
    hsv_wheel[circle_mask, 0] = angles_deg[circle_mask] / 2  # OpenCV  [0,179]
    hsv_wheel[circle_mask, 1] = normalized_radii[circle_mask] * 255  # 
    hsv_wheel[circle_mask, 2] = 255  #  
    
    #   RGB
    hsv_wheel_uint8 = hsv_wheel.astype(np.uint8)
    rgb_wheel = cv2.cvtColor(hsv_wheel_uint8, cv2.COLOR_HSV2RGB)
    
    #        
    center_fade = np.zeros((N, N))
    center_fade[circle_mask] = normalized_radii[circle_mask]
    center_fade = np.stack([center_fade] * 3, axis=-1)
    
    #   (   /)
    rgb_wheel = rgb_wheel.astype(np.float32)
    rgb_wheel = rgb_wheel * center_fade + (1 - center_fade) * 255
    rgb_wheel = np.clip(rgb_wheel, 0, 255).astype(np.uint8)
    
    #   
    background = np.ones((N, N, 3), dtype=np.uint8) * 255
    background[circle_mask] = rgb_wheel[circle_mask]
    
    #     DPI
    fig, ax = plt.subplots(figsize=(10, 10), dpi=100)
    fig.patch.set_facecolor('white')
    
    #   
    extent = [-1, 1, -1, 1]
    ax.imshow(background, extent=extent, origin='upper', interpolation='bilinear')
    
    #   
    #        
    point_sizes = v_filtered * 2 + 0.5  #    
    
    #  scatter plot       
    scatter = ax.scatter(x, y, 
                        c='white', 
                        s=point_sizes, 
                        alpha=0.9,
                        edgecolors='black',
                        linewidths=0.1,
                        rasterized=True)  #   
    
    #  
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal')
    
    #       
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    #     
    circle_border = Circle((0, 0), 1, fill=False, color='lightgray', 
                          linestyle='-', alpha=0.5, linewidth=1)
    ax.add_patch(circle_border)
    
    # 
    if title:
        ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    
    plt.tight_layout()
    
    return fig, ax

def create_broadcast_vectorscope(image_bgr, title='Broadcast Vectorscope'):
    """
         
    """
    #   
    hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
    h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
    s = hsv[:, :, 1].flatten() / 255.0
    v = hsv[:, :, 2].flatten() / 255.0
    
    # 
    valid_mask = (s > 0.02) & (v > 0.05)
    h_filtered = h[valid_mask]
    s_filtered = s[valid_mask]
    v_filtered = v[valid_mask]
    
    # 
    theta = np.deg2rad(h_filtered)
    x = s_filtered * np.cos(theta)
    y = s_filtered * np.sin(theta)
    
    #  
    fig, ax = create_professional_vectorscope(image_bgr, title)
    
    #    broadcast-
    broadcast_targets = [
        #    (75% )
        (0.75 * np.cos(np.deg2rad(0)), 0.75 * np.sin(np.deg2rad(0)), 'R', 'red'),      # Red
        (0.75 * np.cos(np.deg2rad(60)), 0.75 * np.sin(np.deg2rad(60)), 'Yl', 'yellow'), # Yellow
        (0.75 * np.cos(np.deg2rad(120)), 0.75 * np.sin(np.deg2rad(120)), 'G', 'green'), # Green
        (0.75 * np.cos(np.deg2rad(180)), 0.75 * np.sin(np.deg2rad(180)), 'Cy', 'cyan'), # Cyan
        (0.75 * np.cos(np.deg2rad(240)), 0.75 * np.sin(np.deg2rad(240)), 'B', 'blue'),  # Blue
        (0.75 * np.cos(np.deg2rad(300)), 0.75 * np.sin(np.deg2rad(300)), 'Mg', 'magenta') # Magenta
    ]
    
    #   
    for x_pos, y_pos, label, color in broadcast_targets:
        #  
        target_circle = Circle((x_pos, y_pos), 0.05, fill=False, 
                             color='black', linestyle='-', linewidth=2, alpha=0.8)
        ax.add_patch(target_circle)
        
        # 
        ax.text(x_pos * 1.15, y_pos * 1.15, label, ha='center', va='center',
                fontsize=10, fontweight='bold', color='black',
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.9))
    
    #    (25%, 50%, 75%, 100%)
    for radius in [0.25, 0.5, 0.75, 1.0]:
        circle = Circle((0, 0), radius, fill=False, color='gray',
                       linestyle=':', alpha=0.4, linewidth=0.8)
        ax.add_patch(circle)
    
    return fig, ax

# 
if __name__ == "__main__":
    #  
    img_path = 'fotki_for_viktorskop/1.jpg'
    image = cv2.imread(img_path)
    
    if image is not None:
        #   
        fig1, ax1 = create_professional_vectorscope(image, 'Professional Vectorscope')
        plt.show()
        
        #  broadcast 
        fig2, ax2 = create_broadcast_vectorscope(image, 'Broadcast Vectorscope')
        plt.show()
        
        # 
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
        s = hsv[:, :, 1].flatten() / 255.0
        v = hsv[:, :, 2].flatten() / 255.0
        
        print(f"\n===   ===")
        print(f" : {image.shape}")
        print(f" : {len(h):,}")
        print(f" : {np.mean(s):.3f}")
        print(f" : {np.mean(v):.3f}")
        print(f" : {np.median(h[s > 0.1]):.1f}")
        
    else:
        print(f"   : {img_path}")
        
        #    
        print("  ...")
        
        #    
        test_image = np.zeros((400, 400, 3), dtype=np.uint8)
        
        for i in range(400):
            for j in range(400):
                #   HSV 
                center_x, center_y = 200, 200
                dx = j - center_x
                dy = i - center_y
                distance = np.sqrt(dx**2 + dy**2)
                
                if distance < 180:
                    hue = (np.arctan2(dy, dx) + np.pi) / (2 * np.pi) * 179
                    saturation = min(distance / 180 * 255, 255)
                    value = 200
                    
                    hsv_pixel = np.array([[[hue, saturation, value]]], dtype=np.uint8)
                    rgb_pixel = cv2.cvtColor(hsv_pixel, cv2.COLOR_HSV2BGR)
                    test_image[i, j] = rgb_pixel[0, 0]
        
        fig, ax = create_professional_vectorscope(test_image, 'Test Image Vectorscope')
        plt.show()
```
