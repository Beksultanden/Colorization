```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def create_professional_vectorscope(image_bgr, title='Vectorscope', size=800):
    """
    Создает профессиональный векторскоп с гладким цветовым кругом
    """
    # Переводим BGR в HSV
    hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
    
    # Исправляем диапазон H: [0,179] → [0,360]
    h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
    s = hsv[:, :, 1].flatten() / 255.0
    v = hsv[:, :, 2].flatten() / 255.0
    
    # Фильтруем пиксели с низкой насыщенностью и яркостью
    valid_mask = (s > 0.05) & (v > 0.1)
    h_filtered = h[valid_mask]
    s_filtered = s[valid_mask]
    v_filtered = v[valid_mask]
    
    # Полярные координаты
    theta = np.deg2rad(h_filtered)
    x = s_filtered * np.cos(theta)
    y = s_filtered * np.sin(theta)
    
    # Создаем цветовое колесо высокого разрешения
    N = size
    center = N // 2
    radius = center - 20  # отступ от края
    
    # Создаем координатные сетки
    y_coords, x_coords = np.meshgrid(np.arange(N), np.arange(N), indexing='ij')
    dx = x_coords - center
    dy = y_coords - center
    distances = np.sqrt(dx**2 + dy**2)
    
    # Маска для круга
    circle_mask = distances <= radius
    
    # Вычисляем углы (в радианах, затем в градусы)
    angles_rad = np.arctan2(dy, dx)
    angles_deg = np.degrees(angles_rad)
    angles_deg = (angles_deg + 360) % 360  # приводим к [0, 360)
    
    # Нормализованные радиусы (насыщенность)
    normalized_radii = np.minimum(distances / radius, 1.0)
    
    # Создаем HSV колесо
    hsv_wheel = np.zeros((N, N, 3), dtype=np.float32)
    
    # Заполняем только внутри круга
    hsv_wheel[circle_mask, 0] = angles_deg[circle_mask] / 2  # OpenCV использует [0,179]
    hsv_wheel[circle_mask, 1] = normalized_radii[circle_mask] * 255  # насыщенность
    hsv_wheel[circle_mask, 2] = 255  # максимальная яркость
    
    # Конвертируем в RGB
    hsv_wheel_uint8 = hsv_wheel.astype(np.uint8)
    rgb_wheel = cv2.cvtColor(hsv_wheel_uint8, cv2.COLOR_HSV2RGB)
    
    # Создаем градиент к центру для более плавного перехода
    center_fade = np.zeros((N, N))
    center_fade[circle_mask] = normalized_radii[circle_mask]
    center_fade = np.stack([center_fade] * 3, axis=-1)
    
    # Применяем градиент (делаем центр более светлым/белым)
    rgb_wheel = rgb_wheel.astype(np.float32)
    rgb_wheel = rgb_wheel * center_fade + (1 - center_fade) * 255
    rgb_wheel = np.clip(rgb_wheel, 0, 255).astype(np.uint8)
    
    # Создаем белый фон
    background = np.ones((N, N, 3), dtype=np.uint8) * 255
    background[circle_mask] = rgb_wheel[circle_mask]
    
    # Создаем фигуру с высоким DPI
    fig, ax = plt.subplots(figsize=(10, 10), dpi=100)
    fig.patch.set_facecolor('white')
    
    # Отображаем цветовое колесо
    extent = [-1, 1, -1, 1]
    ax.imshow(background, extent=extent, origin='upper', interpolation='bilinear')
    
    # Добавляем точки данных
    # Используем разные размеры точек в зависимости от яркости
    point_sizes = v_filtered * 2 + 0.5  # размер зависит от яркости
    
    # Создаем scatter plot с белыми точками и тонкой черной обводкой
    scatter = ax.scatter(x, y, 
                        c='white', 
                        s=point_sizes, 
                        alpha=0.9,
                        edgecolors='black',
                        linewidths=0.1,
                        rasterized=True)  # для лучшего рендеринга
    
    # Настройки осей
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal')
    
    # Убираем все элементы интерфейса для чистого вида
    ax.set_xticks([])
    ax.set_yticks([])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    # Добавляем тонкую рамку вокруг круга
    circle_border = Circle((0, 0), 1, fill=False, color='lightgray', 
                          linestyle='-', alpha=0.5, linewidth=1)
    ax.add_patch(circle_border)
    
    # Заголовок
    if title:
        ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    
    plt.tight_layout()
    
    return fig, ax

def create_broadcast_vectorscope(image_bgr, title='Broadcast Vectorscope'):
    """
    Создает векторскоп в стиле профессионального видеооборудования
    """
    # Основная обработка изображения
    hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
    h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
    s = hsv[:, :, 1].flatten() / 255.0
    v = hsv[:, :, 2].flatten() / 255.0
    
    # Фильтрация
    valid_mask = (s > 0.02) & (v > 0.05)
    h_filtered = h[valid_mask]
    s_filtered = s[valid_mask]
    v_filtered = v[valid_mask]
    
    # Координаты
    theta = np.deg2rad(h_filtered)
    x = s_filtered * np.cos(theta)
    y = s_filtered * np.sin(theta)
    
    # Создаем векторскоп
    fig, ax = create_professional_vectorscope(image_bgr, title)
    
    # Добавляем профессиональные метки broadcast-стандартов
    broadcast_targets = [
        # Стандартные цветовые полосы (75% насыщенность)
        (0.75 * np.cos(np.deg2rad(0)), 0.75 * np.sin(np.deg2rad(0)), 'R', 'red'),      # Red
        (0.75 * np.cos(np.deg2rad(60)), 0.75 * np.sin(np.deg2rad(60)), 'Yl', 'yellow'), # Yellow
        (0.75 * np.cos(np.deg2rad(120)), 0.75 * np.sin(np.deg2rad(120)), 'G', 'green'), # Green
        (0.75 * np.cos(np.deg2rad(180)), 0.75 * np.sin(np.deg2rad(180)), 'Cy', 'cyan'), # Cyan
        (0.75 * np.cos(np.deg2rad(240)), 0.75 * np.sin(np.deg2rad(240)), 'B', 'blue'),  # Blue
        (0.75 * np.cos(np.deg2rad(300)), 0.75 * np.sin(np.deg2rad(300)), 'Mg', 'magenta') # Magenta
    ]
    
    # Добавляем целевые точки
    for x_pos, y_pos, label, color in broadcast_targets:
        # Целевая область
        target_circle = Circle((x_pos, y_pos), 0.05, fill=False, 
                             color='black', linestyle='-', linewidth=2, alpha=0.8)
        ax.add_patch(target_circle)
        
        # Метка
        ax.text(x_pos * 1.15, y_pos * 1.15, label, ha='center', va='center',
                fontsize=10, fontweight='bold', color='black',
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.9))
    
    # Добавляем концентрические окружности (25%, 50%, 75%, 100%)
    for radius in [0.25, 0.5, 0.75, 1.0]:
        circle = Circle((0, 0), radius, fill=False, color='gray',
                       linestyle=':', alpha=0.4, linewidth=0.8)
        ax.add_patch(circle)
    
    return fig, ax

# Использование
if __name__ == "__main__":
    # Загружаем изображение
    img_path = 'fotki_for_viktorskop/1.jpg'
    image = cv2.imread(img_path)
    
    if image is not None:
        # Создаем профессиональный векторскоп
        fig1, ax1 = create_professional_vectorscope(image, 'Professional Vectorscope')
        plt.show()
        
        # Создаем broadcast векторскоп
        fig2, ax2 = create_broadcast_vectorscope(image, 'Broadcast Vectorscope')
        plt.show()
        
        # Статистика
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        h = hsv[:, :, 0].flatten() * (360.0 / 179.0)
        s = hsv[:, :, 1].flatten() / 255.0
        v = hsv[:, :, 2].flatten() / 255.0
        
        print(f"\n=== Статистика изображения ===")
        print(f"Размер изображения: {image.shape}")
        print(f"Всего пикселей: {len(h):,}")
        print(f"Средняя насыщенность: {np.mean(s):.3f}")
        print(f"Средняя яркость: {np.mean(v):.3f}")
        print(f"Доминирующий оттенок: {np.median(h[s > 0.1]):.1f}°")
        
    else:
        print(f"Не удалось загрузить изображение: {img_path}")
        
        # Демонстрация на тестовом изображении
        print("Создаем тестовое изображение...")
        
        # Создаем градиентное тестовое изображение
        test_image = np.zeros((400, 400, 3), dtype=np.uint8)
        
        for i in range(400):
            for j in range(400):
                # Создаем радиальный HSV градиент
                center_x, center_y = 200, 200
                dx = j - center_x
                dy = i - center_y
                distance = np.sqrt(dx**2 + dy**2)
                
                if distance < 180:
                    hue = (np.arctan2(dy, dx) + np.pi) / (2 * np.pi) * 179
                    saturation = min(distance / 180 * 255, 255)
                    value = 200
                    
                    hsv_pixel = np.array([[[hue, saturation, value]]], dtype=np.uint8)
                    rgb_pixel = cv2.cvtColor(hsv_pixel, cv2.COLOR_HSV2BGR)
                    test_image[i, j] = rgb_pixel[0, 0]
        
        fig, ax = create_professional_vectorscope(test_image, 'Test Image Vectorscope')
        plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb, rgb_to_hsv

# 1) загрузка изображения (RGB в [0..1])
img = iio.imread('fotki_for_viktorskop/1.jpg')[..., :3] / 255.0
hsv = rgb_to_hsv(img)
h, s = hsv[..., 0].ravel(), hsv[..., 1].ravel()

# 2) фон: HSV color wheel
N = 800
y, x = np.ogrid[-1:1:N*1j, -1:1:N*1j]
r = np.sqrt(x**2 + y**2)
theta = (np.arctan2(-y, x) + np.pi) / (2*np.pi)   # [0..1] по углу
wheel = np.zeros((N, N, 3), float)
wheel[..., 0] = theta
wheel[..., 1] = np.clip(r, 0, 1)
wheel[..., 2] = 1.0
wheel_rgb = hsv_to_rgb(wheel)
wheel_rgb[r > 1] = 1.0  # белый вне круга

# 3) рисуем
plt.figure(figsize=(6,6))
plt.imshow(wheel_rgb, extent=(-1,1,-1,1))
# координаты точек по (cos(2πh)*s, sin(2πh)*s)
X = np.cos(2*np.pi*h) * s
Y = np.sin(2*np.pi*h) * s
plt.scatter(X, Y, s=1, c='white', alpha=0.6)
plt.axis('off')
plt.title('HS Color Wheel + Distribution')
plt.show()
```
