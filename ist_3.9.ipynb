```python
from os.path import join, isfile, isdir
from os import listdir
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
from argparse import ArgumentParser

import detectron2
from detectron2.utils.logger import setup_logger
setup_logger()

import numpy as np
import cv2

# import some common detectron2 utilities
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg

import torch
# код для загрузки модели с предобученными весами:При вызове предобученной модели, например, через detectron2.model_zoo, веса автоматически скачиваются с официальных серверов (например, из Model Zoo).
#model_zoo Хранилище моделей:
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_X_101_32x8d_FPN_3x.yaml")) #cfg.merge_from_file(...) Метод merge_from_file загружает предопределенные настройки модели из конфигурационного файла, указанного в аргументе.
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7 #  Устанавливает порог уверенности
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_X_101_32x8d_FPN_3x.yaml") #Здесь model_zoo.get_checkpoint_url() автоматически скачает веса модели при их отсутствии.

#Создается объект DefaultPredictor, который позволяет легко выполнять предсказания с использованием настроенной модели.
predictor = DefaultPredictor(cfg)
```

```python
input_dir = "example" #Устанавливается имя директории, где хранятся исходные файлы. В данном случае это папка "example".

# listdir(input_dir) возвращает список всех элементов в указанной директории (как файлов, так и папок).
# join(input_dir, f) формирует полный путь к каждому элементу, чтобы можно было проверить его тип.
# isfile(...) проверяет, является ли элемент файлом (а не папкой).
# В итоге image_list будет содержать только файлы из директории input_dir.

image_list = [f for f in listdir(input_dir) if isfile(join(input_dir, f))]

# Создается имя выходной директории, добавляя к имени входной директории суффикс "_bbox".
output_npz_dir = "{0}_bbox".format(input_dir)

# Проверяется, существует ли директория с именем output_npz_dir.
# Если директория не существует, выводится сообщение о создании пути.
if os.path.isdir(output_npz_dir) is False:
    print('Create path: {0}'.format(output_npz_dir))
    # Если директория отсутствует, она создается, а программа сообщает об этом.
    os.makedirs(output_npz_dir)
    #Если директория уже существует, ничего не делается.
```

```python
for image_path in image_list: #Цикл, который проходит по каждому элементу списка image_list (то есть по именам файлов изображений в директории).

    img = cv2.imread(join(input_dir, image_path)) # Загружает изображение, используя OpenCV  #join(input_dir, image_path) формирует полный путь к файлу  #cv2.imread() считывает изображение как массив пикселей в формате BGR.
    lab_image = cv2.cvtColor(img, cv2.COLOR_BGR2LAB) # Конвертирует изображение из цветового пространства BGR в цветовое пространство LAB:

    l_channel, a_channel, b_channel = cv2.split(lab_image) # Разделяет LAB-изображение на три отдельных канала

    l_stack = np.stack([l_channel, l_channel, l_channel], axis=2) # Создает массив из L-канала, копируя его трижды вдоль оси каналов:
    outputs = predictor(l_stack) # Передает подготовленное изображение (состоящее из L-канала) в модель через predictor: Модель выполняет предсказание: определяет объекты, их ограничивающие рамки и вероятности.

    #Формирует путь для сохранения результатов
    #image_path.split('.')[0] берет имя файла без расширения (например, для "image1.jpg" это будет "image1").
    #join(output_npz_dir, ...) формирует путь для файла с результатами внутри директории output_npz_dir.
    save_path = join(output_npz_dir, image_path.split('.')[0])

    pred_bbox = outputs["instances"].pred_boxes.to(torch.device('cpu')).tensor.numpy() #Извлекает ограничивающие рамки (bounding boxes) объектов # outputs["instances"].pred_boxes: содержит предсказанные рамки.# .to(torch.device('cpu')): переносит данные на CPU (в случае, если модель работает на GPU).# .tensor.numpy(): преобразует тензор PyTorch в массив NumPy.

   #Извлекает вероятности (scores) предсказанных объектов: # outputs["instances"].scores: содержит вероятности для каждой рамки.
   #.cpu(): переносит данные на CPU. # .data.numpy(): преобразует в массив NumPy.
    pred_scores = outputs["instances"].scores.cpu().data.numpy()

# Сохраняет результаты в формате .npz
# np.savez: сохраняет данные в виде архива NumPy. # save_path: путь для сохранения файла.
# bbox=pred_bbox: сохраняет массив ограничивающих рамок.  # scores=pred_scores: сохраняет массив вероятностей.
    np.savez(save_path, bbox = pred_bbox, scores = pred_scores)

```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Путь к папке с изображениями
folder_path = '/home/user/InstColorization-master/results/'

# Список всех файлов в папке
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.png')]

# Количество изображений
num_images = len(image_files)

# Создаем сетку для отображения изображений
fig, axes = plt.subplots(1, num_images, figsize=(15, 5))  # 1 строка, num_images столбцов

# Если изображение одно, axes будет не массивом, а объектом
if num_images == 1:
    axes = [axes]

# Перебираем изображения и отображаем их
for i, ax in enumerate(axes):
    # Загрузка изображения
    image = cv2.imread(image_files[i])
    # Преобразование из BGR (OpenCV) в RGB (Matplotlib)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Отображение изображения
    ax.imshow(image)
    ax.axis('off')  # Скрыть оси
    ax.set_title(f'Image {i+1}')  # Заголовок для каждого изображения

plt.tight_layout()  # Автоматическая настройка отступов
plt.show()
```
